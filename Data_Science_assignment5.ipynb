### Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([0,5],[0,5])
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Velocity')
plt.show()
```

```python
plt.plot([0,20,40],[20,40,0],color = 'b')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot([0,20,40],[40,0,30],color = 'g')
plt.legend(['Velocity','speed'])
plt.show()
```

```python
plt.plot([0,20,40],[20,40,0],color = 'b' ,ls = ':')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot([0,20,40],[40,0,30],color = 'r',ls = 'dotted' ,linewidth =3)
plt.legend(['Velocity','speed'])
plt.show()
```

```python
plt.plot([1,4,5,6,7],[2,6,3,6,3],'r:o',mfc ='b' ,mec = 'b')
```

### Write a Python programming to display a bar chart of the popularity of programming Languages.
### Sample data: Programming languages: Java, Python, PHP, JavaScript, C#, C++ Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
plt.pie([22.2,17.6,8.8,8,7.7,6.7],labels = ["Java","Python","PHP","JavaScript","C#","C++"],autopct = '%1.1f%%')
plt.show()
```

### Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
### Sample data: Programming languages: Java, Python, PHP, JavaScript, C#, C++ Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
plt.barh([22.2,17.6,8.8,8,7.7,6.7],["Java","Python","PHP","JavaScript","C#","C++"],height =0.1)
plt.show()
```

Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data: Means (men) = (22, 30, 35, 35, 26) Means (women) = (25, 32, 30, 35, 29)

```python
groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

fig, ax = plt.subplots()
index = np.arange(groups)
bar_width = 0.2

rects1 = plt.bar(index, men_means, bar_width, color='b',label='Men')
rects2 = plt.bar(index + bar_width, women_means, bar_width,color='m',label='Women')


plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('Group1', 'Group2', 'Group3', 'Group4', 'Ggroup5'))
plt.legend()
plt.show()
```
